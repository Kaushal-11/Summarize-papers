```python
# Import necessary libraries
import PyPDF2
from nltk.tokenize import sent_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

# Function to perform extractive summarization
def extractive_summary(text, num_sentences=3):
    # Tokenize the text into sentences
    sentences = sent_tokenize(text)
    
    # Create a CountVectorizer to convert sentences to a matrix of token counts
    cv = CountVectorizer().fit_transform(sentences)
    vectors = cv.toarray()
    
    # Compute cosine similarity matrix
    cosine_sim = cosine_similarity(vectors)
    
    # Rank sentences based on their similarity
    sentence_scores = np.sum(cosine_sim, axis=1)
    
    # Get the indices of the top sentences
    top_sentence_indices = np.argsort(sentence_scores)[-num_sentences:]
    
    # Sort the indices
    top_sentence_indices.sort()
    
    # Create the summary
    summary = [sentences[i] for i in top_sentence_indices]
    return ' '.join(summary)

# Path to the PDF file
pdf_file_path = 'D:/SOH/ML_SoH Paper/1.4042987.pdf'

# Extract text from the PDF
pdf_text = extract_text_from_pdf(pdf_file_path)

# Generate extractive summary
summary = extractive_summary(pdf_text, num_sentences=5)

# Print the summary
print("Extractive Summary:")
print(summary)
```
